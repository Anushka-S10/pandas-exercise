```python
# Importing Pandas as a Module
import pandas as pd
import numpy as np
```

```python
# Creates a Dataframe using Pandas

df=pd.DataFrame(np.arange(30).reshape(5,6),index=["Row 1","Row 2","Row 3","Row 4","Row 5"],columns=["Column 1","Column 2","Column 3","Column 4","Column 5","Column 6"])
```

```python
# .head() function returns the Top Rows of the Dataset

df.head()
```

```python
# Access the Row
df.loc["Row 1"]
```

```python
# Accessing the Datatype of the Row
print(type(df))
```

```python
# Access the Column
print(df["Column 1"])
```

```python
# Creating a New Column
df["Column 7"]=df["Column 1"]+df["Column 2"]+df["Column 3"]
df.head()
```

```python
# A Code to calculate Summary Statistics
df.describe()
```

```python

# Calculate the Shape of the Dataframe
df.shape
```

```python
# Calculate the Data Type of Each Column
df.dtypes
```

```python
# Renaming Columns in a Pandas Dataframe
df.columns
```

```python
df.rename(columns={'Column 1':'Col 1','Column 2':'Col 2','Column 3':'Col 3'},inplace="True")
df.head()
```

```python
# We can also rename our Columns using this Trick!
col=["A","B","C","D","E","F","G"]
df.columns=col
df.head()
```
