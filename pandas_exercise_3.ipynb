```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('https://people.sc.fsu.edu/~jburkardt/data/csv/hw_25000.csv')
df.head()
     
```

```python
df.columns
```

```python
col=["Index","Height","Weight"]
df.columns=col
df.head()
```

```python
# Find all the Rows based on any condition in the Column
df.loc[df.Weight>145]
```

```python
# Find all the Rows with more than One Column

df.loc[(df.Weight>150) & (df.Height>60)]
```

```python
# Selecting Required Columns with a condition

df.loc[(df.Weight>=151),["Index","Height"]]
```

```python

# We can select Rows using iloc

df.iloc[[0,10]]
```

```python
# We can also select Rows and Specific Columns

df.iloc[[5,7],[0,2]]
     
```

```python

# We can also select a range using iloc

df.iloc[1:56]
```

```python
# We can also select a Range over Rows and Columns using iloc

df.iloc[52:96,0:2]
```
