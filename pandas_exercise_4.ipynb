```python
# Sometimes, we might encounter Missing Values in the Dataset. So we will be handling them using Pandas.

dataset=pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/Chronic%20Kidney%20Disease%20(CKD)%20Dataset/ChronicKidneyDisease.csv") 
dataset.head()
```

```python
# Checking for NULL values in our dataset

dataset.isnull().values.any()
```

```python
#That's a pity! We have now to clean the dataset to get accurate results of our Data Analysis.

dataset=dataset.dropna(axis = 0, how ='any') 
dataset.isnull().values.any()
```

```python

# Let's check if our Dataset has any NULL values in the Dataset

dataset.isnull().values.any()
     
```

```python
# We can also use other functions like notnull() to check if there are any NULL values in the Dataset. So let's load it for another time.

dataset1=pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/Chronic%20Kidney%20Disease%20(CKD)%20Dataset/ChronicKidneyDisease.csv") 
pd.notnull(dataset1)
```

```python
# Instead of just dropping the Rows where the values are missing, we can handle them using other functions.

dataset1.fillna(0) # It will just replace all NULL values with Zero
```

```python
dataset1.isnull().values.any()
```
