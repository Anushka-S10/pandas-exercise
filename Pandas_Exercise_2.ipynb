```python
import numpy as np
import pandas as pd
```

```python
# Read a CSV File using Pandas
df=pd.read_csv('https://people.sc.fsu.edu/~jburkardt/data/csv/zillow.csv')
df.head()
```

```python
# Let's say, we want to remove the Index Column from our Dataframe
df.drop('Index',axis=1,inplace=True)
df.head()
     
```

```python
# Let's say we want to remove multiple Rows at once. Remember axis=0 refers to Rows while axis=1 refer to Columns
df.drop([0,2], axis=0, inplace=True)
df.head()
```

```python
# Lets sort our Dataset
df1=pd.read_csv('https://people.sc.fsu.edu/~jburkardt/data/csv/grades.csv')
df1.head()
     
```

```python

col=["Last Name","First Name","SSN","Test1","Test2","Test3","Test4","Final","Grade"]
df1.columns=col
df1.head()
```

```python
# In Ascending Order
df1.Final.sort_values().head()
```

```python
# In Descending Order
df1.Final.sort_values(ascending=False).head()
```
